```python
statistics_file = './statistics/cpu_times.csv' #Too long names dosnt work on subprocess (dont know why)

script_name = './run-benchmark.sh'
script_args = " >> " + statistics_file
run_command = script_name + script_args
run_command
```

Run the script to generate all the execution times statistics:

```python
#import os
##os.execv('/bin/sh',[script_name,script_args])
import subprocess
subprocess.call("rm " + statistics_file, shell=True)
subprocess.call(run_command, shell=True)
```
